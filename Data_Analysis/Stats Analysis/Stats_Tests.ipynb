```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
file1 = "../../Data_Collection/Census_API/sample.csv"
file2 = "../../Data_Collection/Census_API/CensusDataFrameClean.csv"

sample = pd.read_csv(file1)
tot_pop = pd.read_csv(file2)
```

```python
sample.head()
```

```python
tot_pop.head()
```

```python
sample["County"] = sample["County"] + ", " + sample["State"]
sample["County"] = sample["County"].str.replace(" County","")
sample.head()
```

```python
sample.drop(columns="State")
sample.head()
```

```python
sample_stat = pd.DataFrame({"County":sample["County"],"Population":sample["Population"]})
tot_pop_stat = pd.DataFrame({"County":tot_pop["Name"],"Population":tot_pop["Population"]})
```

```python
sample_stat.head()
```

```python
tot_pop_stat.head()
```

```python
sample_ev = sample_stat["Population"].mean()
tot_pop_ev = tot_pop_stat["Population"].mean()
```

```python
sample_ev
```

```python
tot_pop_ev
```

```python
tot_pop_stat["Expected Population"] = tot_pop_ev
sample_stat["Expected Population"] = sample_ev
```

```python
crit_val = stats.chi2.ppf(q = 0.95, df = 2)
crit_val
```

```python
stats.chisquare(sample_stat["Population"], sample_stat['Expected Population'])
```

```python
stats.ttest_ind(sample_stat["Population"], tot_pop_stat["Population"], equal_var=False)
```

We are 95% confident that our sample population is statistically similar to the entire population.

p-value = 0.223572
p-value > .05
